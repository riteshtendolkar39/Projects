```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset with appropriate encoding to avoid character errors
df = pd.read_csv("Sample_Superstore.csv", encoding="ISO-8859-1")

# Convert 'Order Date' column to datetime format for time-based analysis
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create a new column 'Total Sales' by multiplying quantity sold by sales amount
df['Total Sales'] = df['Quantity'] * df['Sales']

# Group data by month and calculate total sales for each month
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Total Sales'].sum()

# Set figure size for better visibility
plt.figure(figsize=(12, 8))

# Plot monthly total sales as a bar chart
monthly_sales.plot(kind='bar', title='Monthly Sales')

# Display the plot
plt.show()
```
